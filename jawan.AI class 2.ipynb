```python
eng = int(input("enter english marks"))
isl = int(input("enter isl marks:"))
math = int(input("enter math marks:"))
urdu = int(input("enter urdu marks:"))
comp = int(input("enter comp marks:"))

obtain = eng + isl + math + urdu + comp 

print("your obtain marks is : ",obtain)

perc = obtain / 500 * 100

print("your percentage is : ",perc)

if perc >= 100 and perc <= 80:
    print("A1")
elif perc >= 70 and perc < 80:
    print("A grade")
elif perc >= 60 and perc < 70:
    print("b")
elif perc >= 50 and perc < 50:
    print("c")
elif perc >= 40 and perc < 50:
    print("d")
elif perc >= 33 and perc < 40:
    print("e")   
else:
    print("fail")
```
