```python

##### Marksheet with grades

eng = int(input("enter english marks"))
isl = int(input("enter isl marks:"))
math = int(input("enter math marks:"))
urdu = int(input("enter urdu marks:"))
comp = int(input("enter comp marks:"))
phys = int(input("enter phys marks:"))

obtain = eng + isl + math + urdu + comp + phys

print("enter obtain mark is :", obtain)

perc = obtain / 500 * 100

print("enter percentage :" ,perc)

if perc >= 100 and perc <= 80:
   print("a")
elif perc >= 70 and perc <= 60:
    print("b")
elif perc >= 60 and perc <= 40:
    print("c")
elif perc >= 40 and perc <= 33:
     print("d")
else:
     print("f")

```
