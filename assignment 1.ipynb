```python
input1 = int(input("enter value 1:"))
input2 = int(input("enter value 2:"))
operand = int(input("enter value operand"))

if operand == "+":
    print(input1 + input2)
elif operand == "-":
    print(input1 - input2)
elif operand == "/":
    print(input1 / input2)
elif operand == "*":
    print(input1 * input2)
else:
    print("please insert correr operator")
```
